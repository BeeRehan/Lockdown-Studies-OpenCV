# cv2 with Matplotlib

```python
import cv2
```

```python
img = cv2.imread(r'C:\Users\MD\Documents\Python Scripts\lockdown #1\src\img\MD.jpg')
plt.imshow(img) #Matpoltlib and openCV use different color channel (i.e) plt=RGB && cv2=BGR 
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(fix_img)
```

```python
gray_img = cv2.imread(r'C:\Users\MD\Documents\Python Scripts\lockdown #1\src\img\MD.jpg',0)
plt.imshow(gray_img,cmap='gray'), gray_img.shape
```

# Resize

```python
#Normal Resize
res_img1 = cv2.resize(fix_img,(1024,720))#(width,height)
plt.imshow(res_img1)
```

```python
#resize with ratio
width = 0.7
height = 0.7
res_img2 = cv2.resize(img,(0,0),img,width,height)
plt.imshow(res_img2)
```

```python
cv2.imwrite('saves\saveMD.jpg',res_img2)
```
