# Drawing shapes on images

```python
img = np.zeros((600,600,3),int)
plt.imshow(img)
```

```python
#Draw rectangle

rect = cv2.rectangle(img,pt1=(400,100),pt2=(550,300),color=(0,0,255),thickness=10)
plt.imshow(rect)
```

```python
#Draw circle

circ = cv2.circle(img,center=(200,200),radius=50,color=(255,0,0),thickness=10)
plt.imshow(circ)
```

```python
#Draw line

line1 = cv2.line(img,pt1=(0,400),pt2=(200,600),color=(0,0,255),thickness=1)
plt.imshow(line1)
```

```python
line2 = cv2.line(img,(400,600),(600,400),(0,0,255),1)
plt.imshow(line2)
```

```python
#Draw custom polygon
ver = np.array([[250,350], [300,400],
               [350,350], [300,580]],int)
ver.shape

rever = ver.reshape(-1,1,2)

rever.shape
poly = cv2.polylines(img,[rever],isClosed=True,color=(0,255,0),thickness=10)
plt.imshow(poly)
```

# Drawing using Mouse Event

```python
import cv2
import numpy as np
```

```python

def draw(event,x,y,flags,param):
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.rectangle(img,(x,y),(x+50,y+50),(255,31,255),-1)
        
cv2.namedWindow(winname="Drawing")
cv2.setMouseCallback("Drawing",draw)

img = np.zeros((1024,720,3),np.int8)

while True:
    cv2.imshow("Drawing",img)
    if cv2.waitKey(1) & 0xFF==27:
        break
cv2.destroyAllWindows()
```

# Drawing with movement

```python
status = False
ix,iy = -1,-1

def draw_move(event,x,y,flags,params):
    
    global ix,iy,status
    
    if event==cv2.EVENT_LBUTTONDOWN:
        
        status = True
        ix,iy = x,y
    
    elif event==cv2.EVENT_MOUSEMOVE:
        if status == True:
            cv2.rectangle(img,(ix,iy),(x,y),(133,0,133),-1)
    elif event==cv2.EVENT_LBUTTONUP:
        status = False
        cv2.rectangle(img,(ix,iy),(x,y),(31,252,33),1)

cv2.namedWindow("L2R")
cv2.setMouseCallback("L2R",draw_move)

img = np.zeros((1070,790,3))

#cap = cv2.VideoCapture(0)
while True:
    
 #   ros, img = cap.read()
    cv2.imshow("L2R",img)
    
    if cv2.waitKey(1) & 0xFF ==27:
        break
        
#cap.release()
cv2.destroyAllWindows()
```

##### Blending an img of same shape

```python
img1 = cv2.cvtColor(cv2.imread("src\Dog.jpg"),cv2.COLOR_BGR2RGB)
#img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img = img1.copy()
plt.imshow(img1);
print("Shape",img1.shape)
```

```python
logo = cv2.imread("src\logo.png")
logo = cv2.cvtColor(logo,cv2.COLOR_BGR2RGB)
plt.imshow(logo);
print("Shape",logo.shape)
```

```python
img1 = cv2.resize(img1,(1300,1900))

logo = cv2.resize(logo,(1300,1900))

img1.shape, logo.shape
```

```python
plt.imshow(img1) 
```

```python
plt.imshow(logo)
```

```python
blended = cv2.addWeighted(img1,0.9,logo,0.2,0.3)
plt.imshow(blended)
```
